```python
import pandas as pd

df = pd.read_csv("../data/Superstore.csv", encoding="latin1")
df.head()
```

```python
print(df.shape)
print(df.columns.tolist())
print(df.info())
print(df.isnull().sum())
print(df.duplicated().sum())
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("-", "_")
)

df.columns.tolist()
```

```python
df["order_date"] = pd.to_datetime(df["order_date"], errors="coerce")
df["ship_date"] = pd.to_datetime(df["ship_date"], errors="coerce")
```

```python
df = df.drop_duplicates()
```

```python
df.isnull().sum()
```

```python
df["year"] = df["order_date"].dt.year
df["month"] = df["order_date"].dt.month
df["month_name"] = df["order_date"].dt.month_name()
df["quarter"] = df["order_date"].dt.quarter
```

```python
df["shipping_days"] = (df["ship_date"] - df["order_date"]).dt.days
```

```python
df["profit_margin"] = df["profit"] / df["sales"]
df["profit_margin"] = df["profit_margin"].replace([float("inf"), -float("inf")], 0)
df["profit_margin"] = df["profit_margin"].fillna(0)
```

```python
df["is_loss"] = (df["profit"] < 0).astype(int)
```

```python
def discount_band(x):
    if x == 0:
        return "No Discount"
    elif x <= 0.2:
        return "Low"
    elif x <= 0.5:
        return "Medium"
    else:
        return "High"

df["discount_band"] = df["discount"].apply(discount_band)
```

```python
df.to_csv("../data/superstore_clean.csv", index=False)
```

```python
df = pd.read_csv("/Users/htetwaiyanoo/Senior_Project/data/superstore_clean.csv")
df.head()
```

```python
df["order_date"] = pd.to_datetime(df["order_date"])
df["year"] = df["order_date"].dt.year
df["month"] = df["order_date"].dt.month
```

```python
monthly_sales = df.groupby(["year", "month"])["sales"].sum().reset_index()

monthly_sales["date"] = pd.to_datetime(
    monthly_sales["year"].astype(str) + "-" +
    monthly_sales["month"].astype(str) + "-01"
)

monthly_sales = monthly_sales.sort_values("date").reset_index(drop=True)

monthly_sales = monthly_sales[["date", "sales"]]

monthly_sales.head()
```

```python
train = monthly_sales.iloc[:-6].copy()
test = monthly_sales.iloc[-6:].copy()

print("Train rows:", len(train))
print("Test rows:", len(test))
print(test)
```

```python
from sklearn.linear_model import LinearRegression

train["time_index"] = range(len(train))
test["time_index"] = range(len(train), len(train) + len(test))

lr_model = LinearRegression()
lr_model.fit(train[["time_index"]], train["sales"])

test["lr_pred"] = lr_model.predict(test[["time_index"]])

test[["date", "sales", "lr_pred"]]
```

```python
all_data = monthly_sales.copy()
all_data["ma3_pred"] = all_data["sales"].rolling(window=3).mean().shift(1)

test["ma3_pred"] = all_data.loc[test.index, "ma3_pred"].values

test[["date", "sales", "lr_pred", "ma3_pred"]]
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

lr_mae = mean_absolute_error(test["sales"], test["lr_pred"])
lr_rmse = np.sqrt(mean_squared_error(test["sales"], test["lr_pred"]))

ma3_mae = mean_absolute_error(test["sales"], test["ma3_pred"])
ma3_rmse = np.sqrt(mean_squared_error(test["sales"], test["ma3_pred"]))

comparison = pd.DataFrame({
    "Model": ["Linear Regression", "3-Month Moving Average"],
    "MAE": [lr_mae, ma3_mae],
    "RMSE": [lr_rmse, ma3_rmse]
})

comparison
```
